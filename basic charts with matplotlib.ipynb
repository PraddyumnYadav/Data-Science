## Mutiple Lines in a single chart and change style

```python
x = [1, 4.5, 6, 8, 10, 15, 20, 21, 23, 30, 31, 32, 33, 34, 35, 36, 40, 42]
x2 = [1, 6, 6.5, 5, 10, 10, 17.5, 18, 3, 3, 4, 5, 6, 7, 8, 1, 25, 25]
y = [1, 4.5, 6, 8, 10, 15, 20, 21, 23, 30, 31, 32, 33, 34, 35, 36, 40, 42]
y2 = [1, 4, 7, 4, 12, 10, 16, 18, 2, 2, 3, 4, 5, 6, 7, 1, 23, 23]

style.use('ggplot')

plt.plot(x,x2, label='Google')
plt.plot(y,y2, label='Apple')
plt.legend()
plt.show()
```
